# Basics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Guni Function
def gini(p):
    return 1-(p**2).sum()
```

```python
class Node:
    def __init__(self, X, y, level):
        self.level = level
        self.data_num = len(y) # number of data
        self.branch = []
        
        # born to be trained 
        if (len(np.unique(y))==1):
            self.y = y[0]
        elif (level==0):
            self.y = np.bincount(y).argmax()
            # bincount: Count number of occurrences of each value in array of non-negative ints.
            # take a value which occured most
        else:
            self.gn = 1 # initialize of gini purity
                        # 0 is pure, 1 is dirty
            
            for j in range(X.shape[1]): # feature
                x = X[:,j]
                xas = x.argsort()
                x_sort = x[xas]
                y_sort = y[xas]

                l = (x_sort[1:] + x_sort[:-1])/2 # maybe reduce l-candidate
                l = l[y_sort[1:] != y_sort[:-1]] # for faster tree building
                
                for th in l:
                    bunki = th>x
                    y_left  = y[bunki]
                    y_right = y[~bunki]
                    n_left  = float(len(y_left))
                    n_right = float(len(y_right))
                    gn = (gini(np.bincount(y_left)/n_left)*n_left
                         +gini(np.bincount(y_right)/n_right)*n_right)/self.data_num
                    
                    if (self.gn > gn): # update gn for getting best gn
                        self.gn = gn
                        self.j  = j  # column of feature
                        self.th = th # th of feature
                        
            o = (self.th > X[:, self.j])
            
            self.branch = [Node(X[o], y[o], level-1), 
                           Node(X[~o], y[~o], level-1)]
            
    def __call__(self, X): # inference
        if (self.branch==[]):
            return self.y
        else:
            o = self.th > X[:, self.j]
            return np.where(o, self.branch[0](X), self.branch[1](X))
            # what is this np.where working for?
            # -> th>X then replace it self.branch[0]
            # -> th<X then replace it self.branch[1]
            # https://qiita.com/natsu-re/items/37761ff441e9e31e2b82
```

```python
class DecisionTree:
    def __init__(self, depth):
        self.depth = depth
        
    def training(self, X, y):
        self.structure = Node(X, y, self.depth)
        
    def predict(self, X):
        return self.structure(X)
```

```python
class RandomForest:
    def __init__(self, depth, n_estimator=10):
        self.depth = depth
        self.n_estimator = n_estimator
        
    def training(self, X, z):
        n = len(z)
        self.k = z.max()+1
        self.estimator_list = []
        
        for i in range(self.n_estimator):
            tmp_estimator = DecisionTree(self.depth)
            input_ind = np.random.choice(n, n)
            tmp_estimator.training(X[input_ind], z[input_ind])
            self.estimator_list.append(tmp_estimator)
            
    def make_predict_list(self, X):
        result = 0
        for tmp_tree in self.estimator_list:
            result += tmp_tree.predict(X)[:, None]==np.arange(self.k)
        return result
    
    def predict(self, X):
        return self.make_predict_list(X).argmax(1)
    
    def predict_proba(self, X):
        return self.make_predict_list(X)/self.n_estimator
```

# Usage of Tree

```python
from sklearn import datasets
X, y = datasets.make_blobs(n_samples=100, n_features=2, centers=5,
                           cluster_std=1.7, random_state=2)
#print(X)     # X is [n_data, 2]
#print(label) # label is [n_data]
```

```python
DTree = DecisionTree(100)
DTree.training(X, y)

mx, my = np.meshgrid(np.linspace(X[:,0].min()-1,X[:,0].max()+1,200),
                     np.linspace(X[:,1].min()-1,X[:,1].max()+1,200))
mX = np.stack([mx, my], 1)
mz = DTree.predict(mX)

plt.gca(aspect=1)
plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k', cmap='plasma')
plt.contourf(mx,my,mz, alpha=0.4, cmap='plasma', zorder=0)
plt.show()
```

```python
mx,my = np.meshgrid(np.linspace(X[:,0].min()-1,X[:,0].max()+1,200),
                    np.linspace(X[:,1].min()-1,X[:,1].max()+1,200))
mX = np.stack([mx,my],1)

plt.figure(figsize=[9,7])
for f in range(1,7):
    DTree = DecisionTree(f)
    DTree.training(X,y)
    mz = DTree.predict(mX)
    plt.subplot(230+f,aspect=1)
    plt.scatter(X[:,0],X[:,1],20,c=y,edgecolor='k',cmap='plasma')
    plt.contourf(mx,my,mz,alpha=0.4,cmap='plasma',zorder=0)
    plt.title('%d'%f)
    
plt.show()
```

# Usage of RandomForest

```python
np.random.seed(2)
X = np.random.uniform(-1,1,[500,2])
z = (np.sqrt(X[:,0]**2+X[:,1]**2)//0.282).astype(int)

Rforest = RandomForest(100)
Rforest.training(X, z)
nmesh = 200
mx,my = np.meshgrid(np.linspace(X[:,0].min(),X[:,0].max(),nmesh),
                    np.linspace(X[:,1].min(),X[:,1].max(),nmesh))
mX = np.stack([mx.ravel(),my.ravel()],1)
mz = Rforest.predict(mX).reshape(nmesh,nmesh)

plt.gca(aspect=1, xlim=[mx.min(),mx.max()], ylim=[my.min(),my.max()])
plt.scatter(X[:,0],X[:,1],alpha=0.6,c=z,edgecolor='k',cmap='rainbow')
plt.contourf(mx,my,mz,alpha=0.4,cmap='rainbow',zorder=0)
plt.show()
```

```python
np.random.seed(11)
X = np.random.uniform(-1,1,[1000,2])
z = (np.sqrt(X[:,0]**2+X[:,1]**2)//0.707).astype(int)

Rforest = RandomForest(100, n_estimator=100)
Rforest.training(X, z)
nmesh = 200
mx,my = np.meshgrid(np.linspace(-1,1,nmesh), np.linspace(-1,1,nmesh))
mX = np.stack([mx.ravel(),my.ravel()],1)
mz = Rforest.predict_proba(mX)[:,1].reshape(nmesh,nmesh)

plt.gca(aspect=1,xlim=[mx.min(),mx.max()],ylim=[my.min(),my.max()])
plt.scatter(X[:,0],X[:,1],c=z,alpha=0.9,edgecolor='k',cmap='Spectral')
plt.contourf(mx,my,mz,100,cmap='Spectral',zorder=0)
plt.colorbar(pad=0.01)
plt.show()
```
